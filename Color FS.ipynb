```python
import streamlit as st
import numpy as np
import pickle

from google.colab import drive
drive.mount('/content/drive')

# Load mô hình và scaler
with open('/content/drive/MyDrive/rf_classifier_model.pkl', 'rb') as f:
    rf_classifier = pickle.load(f)
with open('/content/drive/MyDrive/scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

# Hàm dự đoán
def predict_group(L, a, b):
    sample = np.array([[L, a, b]])
    sample_scaled = scaler.transform(sample)
    prediction = rf_classifier.predict(sample_scaled)
    probabilities = rf_classifier.predict_proba(sample_scaled)
    return prediction[0], probabilities[0]

# Giao diện người dùng với Streamlit
st.title("Dự đoán Bản Chất Mẫu dựa trên L, a, b")

L = st.number_input("25.2", min_value=0.0, max_value=100.0, value=25.99)
a = st.number_input("33.9", min_value=0.0, max_value=100.0, value=33.9)
b = st.number_input("42.35", min_value=0.0, max_value=100.0, value=42.35)

if st.button("Dự đoán"):
    prediction, probabilities = predict_group(L, a, b)
    st.write(f"Dự đoán Bản Chất Mẫu: {prediction}")
    st.write("Xác suất cho các nhóm:")
    for group, prob in zip(rf_classifier.classes_, probabilities):
        st.write(f"{group}: {prob:.2f}")
```
